Importing the dependencies

```python
import numpy as np
import pandas as pd
import sklearn.datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

Data Collection and Processing

```python
# loading the data from sklearn

breast_cancer_dataset=sklearn.datasets.load_breast_cancer()

print(breast_cancer_dataset)
```

```python
# loading the data to a data frame

data_frame=pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows

data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label']=breast_cancer_dataset.target
```

```python
#printing last 5 rows of dataframe
data_frame.tail()
```

```python
# number of rows and columns

data_frame.shape
```

```python
# getting some information about the data
data_frame.info()
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
# statistical measures about the data
data_frame.describe()
```

```python
# checking the distribution of target variables
data_frame['label'].value_counts()
```

1--- Benign  0----Malignant

```python
data_frame.groupby('label').mean()
```

**Separating Target and Features**

```python
X= data_frame.drop(columns='label',axis=1)
Y= data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

**Standardize the data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)   # transform only
```

```python
print(X_test_std)
```

**BUILDING THE NEURAL NETWORK**

```python
# importing Tensorflow and Keras

import tensorflow as tf   # in google colab it's already installed, but need to install if using others
tf.random.set_seed(3)   # almost same with random state
from tensorflow import keras  # Keras---backend of tensorflow
```

```python
# setting up the layers of Neural Network

model= keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),         # input layers # flatten --use to convert data into single dimension
                          keras.layers.Dense(20, activation='relu'),       # hidden layers  # can have other hidden layers too
                          keras.layers.Dense(2,activation='sigmoid')       # output layers    # 2--represents the number of classes that you have, here it is malignant and benign

])
```

```python
# compiling the Neural Networsk

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

Reminder: **loss='sparse_categorical_crossentropy'** when you have 0,1,2 categorical

*   Example    DOG=0, CAT=1, HORSE=2

If more than 3 use  THE
**loss='categorical_crossentropy'**

```python
# Training the neural network

history = model.fit(X_train_std,Y_train, validation_split=0.1,epochs=10)
```

**Visualizing Accuracy and Loss**

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data','validation data'], loc ='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data','validation data'], loc ='upper right')
```

**Accuracy of the model on test data**

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print (accuracy)
```

```python
# Converting the prediction probability to class labels

# argmax will give the maximum value using the index number of that maximum value

Y_pred_labels = [np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```

**Building the Predictive System**

```python
input_data= (13.08,15.71,85.63,520,0.1075,0.127,0.04568,0.0311,0.1967,0.06811,0.1852,0.7477,1.383,14.67,0.004097,0.01898,0.01698,0.00649,0.01678,0.002425,14.5,20.49,96.09,630.5,0.1312,0.2776,0.189,0.07283,0.3184,0.08183)

# change the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# need to reshape first before standardize the data
# standardizing the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction= model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if (prediction_label[0]==0):
  print('The tumor is Malignant')

else:
     print('The tumor is Benign')

```
